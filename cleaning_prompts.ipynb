```python
import json

bad = 0

with open("clean_labeled_dataset.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        try:
            obj = json.loads(line)

            if "prompt" not in obj:
                bad += 1
            if "topics" not in obj:
                bad += 1
            if "intents" not in obj:
                bad += 1

        except:
            bad += 1

print("Broken rows:", bad)
```

```python
import json
from collections import Counter

topic_counter = Counter()
intent_counter = Counter()

with open("clean_labeled_dataset.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)

        for t in obj["topics"]:
            topic_counter[t] += 1

        for i in obj["intents"]:
            intent_counter[i] += 1


print("TOPIC DISTRIBUTION\n")
for k,v in topic_counter.most_common():
    print(k, v)

print("\nINTENT DISTRIBUTION\n")

for k,v in intent_counter.most_common():
    print(k, v)
```

```python
import json
from collections import Counter

matrix = Counter()

with open("clean_labeled_dataset.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)

        for t in obj["topics"]:
            for i in obj["intents"]:
                matrix[(t,i)] += 1


print("TOPIC × INTENT COMBINATIONS\n")

for k,v in matrix.most_common(20):
    print(k, v)
```

```python
import json
from collections import Counter

word_counter = Counter()
# intent_counter = Counter()

with open("clean_labeled_dataset.jsonl", "r", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)
        word_count = len(obj["prompt"].split())
        if word_count <=10:
            bucket = "0-10"
        elif word_count <=20:
            bucket = "11-20"
        elif word_count <=40:
            bucket = "21-40"
        else:
            bucket = "40+"
        word_counter[bucket] += 1

print("WORD COUNT DISTRIBUTION\n")
for k,v in word_counter.most_common():
    print(k, v)
```

```python
import json

INPUT = "clean_labeled_dataset.jsonl"
OUTPUT = "final_dataset.jsonl"

VALID_TOPICS = {
    "technical","academic","career",
    "project","creative","meta_learning","general"
}

VALID_INTENTS = {
    "information_request","implementation_help",
    "troubleshooting","decision_making",
    "strategic_planning","validation_seeking",
    "idea_exploration","constraint_analysis",
    "emotional_processing"
}

seen_prompts = set()
clean_rows = []

with open(INPUT,"r",encoding="utf-8") as f:
    for line in f:

        obj = json.loads(line)
        prompt = obj["prompt"]

        if prompt in seen_prompts:
            continue

        seen_prompts.add(prompt)

        topics = [t for t in obj["topics"] if t in VALID_TOPICS]
        intents = [i for i in obj["intents"] if i in VALID_INTENTS]

        if len(topics)==0 or len(intents)==0:
            continue

        clean_rows.append({
            "prompt":prompt,
            "topics":topics,
            "intents":intents
        })

with open(OUTPUT,"w",encoding="utf-8") as f:
    for row in clean_rows:
        f.write(json.dumps(row)+"\n")

print("Final dataset size:",len(clean_rows))
```
